```python
import pandas as pd
import numpy as np
from patsy import dmatrices
import statsmodels.formula.api as sm
from statsmodels.datasets.longley import load
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.DataFrame(load().exog, columns=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
df['Y'] = load().endog
```

```python
df.head()
```

```python
results = sm.ols(formula='Y~X1+X2+X3+X4+X5+X6', data=df).fit()
```

```python
df_x = df.copy()
df
```

```python
predicted_Y = results.predict(df_x)
```

```python
predicted_Y.head()
```

```python
results.summary()
```

```python
vif = pd.DataFrame()
```

```python
vif["VIF Factor"] = [variance_inflation_factor(df_x.values, i) for i in range(df_x.shape[1])]
vif["features"] = df_x.columns
```

```python
vif.round(1)
```

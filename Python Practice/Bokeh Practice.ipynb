# Warm up steps

## Create simple figure

```python
# Import necessary modules
from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook, reset_output
import numpy as np
import pandas as pd
```

```python
# Generate mock data
np.random.seed(0) # Random seed
epsilon = np.random.normal(0, 200, 2000)
#X = np.log((200) * np.random.random_sample(1000) + 5000)
#Y = (25 * np.exp(X ** 2)) + (-100 * np.exp(X)) + epsilon
X = ((50) * np.random.random_sample(2000)) + 0
Y = (25 * (X ** 2)) + (-100 * X) + epsilon
```

```python
# Figures will be displayed in this jupyter notebook
output_notebook()

# Create a figure
p = figure(x_axis_label='Study time', y_axis_label='Salary')

# Show the figure
show(p)
```

```python
# Clear the last rendered figure
reset_output()
```

```python
output_notebook()
# Define a more advanced figure
p = figure(x_axis_label='Study time (Thousand Hours) - Linear scale ticks  ', \
           y_axis_label='Salary (USD) - Linear scale ticks', \
          x_axis_type='linear',
          y_axis_type='linear',
          title='Just another simple figure',
          x_axis_location='above',
          y_axis_location='left',
          title_location='right',
          toolbar_location='below')

# Render data on figure
p.circle(x=X, y=Y, alpha=0.5)

# Display the figure
show(p)
```

```python
reset_output()
output_notebook()
# Define a more advanced figure
p = figure(x_axis_label='Study time (Thousand Hours) - Log scale ticks  ', \
           y_axis_label='Salary (USD) - Linear scale ticks', \
          x_axis_type='log',
          y_axis_type='linear',
          title='Just another simple figure',
          x_axis_location='above',
          y_axis_location='left',
          title_location='right',
          toolbar_location='below')

# Render data on figure
p.circle(x=X, y=Y, alpha=0.5)

# Display the figure
show(p)
```

```python
reset_output()
output_notebook()
# Define a more advanced figure
p = figure(x_axis_label='Study time (Thousand Hours) - Linear scale ticks  ', \
           y_axis_label='Salary (USD) - Log scale ticks', \
          x_axis_type='linear',
          y_axis_type='log',
          title='Just another simple figure',
          x_axis_location='above',
          y_axis_location='left',
          title_location='right',
          toolbar_location='below')

# Render data on figure
p.circle(x=X, y=Y, alpha=0.5)

# Display the figure
show(p)
```
